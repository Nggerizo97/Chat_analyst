```python
import json

# Paso 1: Cargar el archivo JSON
with open('Conversaci√≥n_2.json', 'r', encoding='utf-8') as file:
    content = file.read()

# Suponiendo que cada objeto JSON est√° separado por nuevas l√≠neas
json_strings = content.split('\n')

data = [json.loads(j) for j in json_strings if j]



# Paso 2: Acceder a los datos
for item in data:
    try:
        if 'result' in item:
            # Parsear el JSON contenido en '_raw' para acceder correctamente a los datos
            raw_data = json.loads(item['result']['_raw'])
            # Acceder a 'request' y luego a 'context'
            chat_history = raw_data['request']['context']
            print(chat_history)
        else:
            print("La clave 'result' no se encuentra en el elemento.")
    except KeyError as e:
        print(f"Error de clave no encontrada: {e}")



```

```python
import pandas as pd
import json

# Inicializar una lista para almacenar el historial de chat
chat_history_list = []

# Paso 2: Acceder a los datos
for item in data:
    try:
        if 'result' in item:
            # Parsear el JSON contenido en '_raw' para acceder correctamente a los datos
            raw_data = json.loads(item['result']['_raw'])
            # Acceder a 'request' y luego a 'context'
            chat_history = raw_data['request']['context']
            # Asumiendo que 'chat_history' puede ser una lista de mensajes
            if isinstance(chat_history, list):
                for message in chat_history:
                    # Agregar cada mensaje a la lista chat_history_list
                    chat_history_list.append(message)
            else:
                # Si no es una lista, agregar el diccionario completo
                chat_history_list.append(chat_history)
        else:
            print("La clave 'result' no se encuentra en el elemento.")
    except KeyError as e:
        print(f"Error de clave no encontrada: {e}")

# Convertir la lista de historial de chat a DataFrame
df1 = pd.DataFrame(chat_history_list)

```

```python
all_messages = []  # Lista principal para almacenar listas de mensajes de cada iteraci√≥n

for f in chat_history_list:
    temp_list = []  # Lista temporal para almacenar mensajes de la iteraci√≥n actual
    for i in f['chatHistory']:
        temp_list.append(i['text'])
    # A√±adir la lista temporal a la lista principal despu√©s de cada iteraci√≥n
    all_messages.append(temp_list)

# Para imprimir todas las listas de mensajes
cantidad_mensajes = []  # Paso 1

for messages in all_messages:  # Paso 2
    cantidad_mensajes.append(len(messages))

df1['cantidad_mensajes'] = cantidad_mensajes  # Paso 3
print(df1['cantidad_mensajes'])



# Paso 4: Guardar el DataFrame en un archivo Excel
df1.to_csv('mi_archivo_excel2.xlsx', index=False)
```
